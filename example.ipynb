```python
import docx
import fitz  # PyMuPDF
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer
```

```python



def read_word_document(docx_path):
    doc = docx.Document(docx_path)
    full_text = []
    for paragraph in doc.paragraphs:
        full_text.append(paragraph.text)
    return "\n".join(full_text)

def read_pdf_document(pdf_path):
    doc = fitz.open(pdf_path)
    full_text = []
    for page_num in range(doc.page_count):
        page = doc[page_num]
        full_text.append(page.get_text())
    return "\n".join(full_text)

def summarize_text(text, num_sentences=3):
    parser = PlaintextParser.from_string(text, Tokenizer("english"))
    summarizer = LsaSummarizer()
    summary = summarizer(parser.document, num_sentences)
    summarized_text = " ".join([str(sentence) for sentence in summary])
    return summarized_text

# Example usage
word_doc_path = "path/to/your/document.docx"
pdf_doc_path = "path/to/your/document.pdf"

word_text = read_word_document(word_doc_path)
pdf_text = read_pdf_document(pdf_doc_path)

word_summary = summarize_text(word_text)
pdf_summary = summarize_text(pdf_text)

print("Word Document Summary:")
print(word_summary)

print("\nPDF Document Summary:")
print(pdf_summary)
```
