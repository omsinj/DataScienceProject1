We start with by importing the necessary libraries. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

Import and inspect the data

```python
# Import the data
df = pd.read_csv("insurance.csv")
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
# Extracting only 'age' and 'charges' columns
subset_df = df[['age', 'charges']]

# Plotting the scatter plot
plt.scatter(subset_df['age'], subset_df['charges'], label='Actual Data', color='blue')

plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot: Age vs Charges')
plt.legend()
plt.show()
```

Task 1 - Plot a scatter plot with age on the x-axis and charges on the y-axis.
Using linear_model.LinearRegressions() from sklearn, fit a model to the data, and make predictions on the data


```python
# Extracting features (X) and target variable (y)
X = df[['age']]
y = df['charges']

# Fit a linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions on the data
y_pred = model.predict(X)

# Plotting the scatter plot, note this plot is without a best fit line.
plt.scatter(X, y, label='Actual Data', color='blue')

plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Scatter Plot: Age vs Charges')
plt.legend()
plt.show()
```

Plotting a scatter plot with a best fit line

```python
# Plotting the scatter plot
plt.scatter(df['age'], df['charges'], label='Actual Data', color='blue')

# Plotting the best-fit line
plt.plot(df['age'], model.predict(df[['age']]), label='Best-Fit Line', color='red')

plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Linear Regression: Age vs Charges with Best-Fit Line')
plt.legend()
plt.show()
```

Creating predictions on the data

```python

# Make predictions on the data
predicted_charges = model.predict(X)

# Plotting the scatter plot with the actual and predicted charges
plt.scatter(X, y, label='Actual Charges', color='blue')
plt.scatter(X, predicted_charges, label='Predicted Charges', color='red', marker='x')

# Plotting the best-fit line
plt.plot(X, predicted_charges, label='Best-Fit Line', color='green')

plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Actual vs Predicted Charges with Best-Fit Line')
plt.legend()
plt.show()
```

Creating a table to show actual vs predicted charges

```python
# import the necessary library
from prettytable import PrettyTable
```

```python
# Creating a PrettyTable
table = PrettyTable(['Age', 'Actual Charges', 'Predicted Charges'])

# Populating the table for the first 5 rows
for i in range(5):
    age, actual, predicted = X['age'][i], round(y[i], 2), round(predicted_charges[i], 2)
    table.add_row([age, actual, predicted])

# Printing the table
print(table)
```
